```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
deaths_2015 = pd.read_csv('mortality/2015_data.csv', usecols=["month_of_death"])
```

```python
plt.figure(figsize=(16, 9))
p = sns.countplot(x='month_of_death', data=deaths_2015, color='#ff6c23');
_ = p.set(xlabel="2015", ylabel="Death Count", 
      xticklabels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
     title= "US Deaths per month in 2015")
_ = sns.set_style("dark")
```
